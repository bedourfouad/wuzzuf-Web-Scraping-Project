```python
df = pd.DataFrame(columns=["job_name", "company_name", "country", "location_name", "job_type", "exe_level", "exe_year", "skills"])
df
```

```python
for i in range(10): 
    request = requests.get(f"https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=data%20science&start={i}")

    content=request.content

    soup=BeautifulSoup(content,"html.parser")

    jobs = soup.find_all("div",class_="css-1gatmva e1v1l3u10")

    for job in jobs:
        job_name=jobs[0].find_all("h2",class_="css-m604qf")[0].text
        company_name=jobs[0].find_all("div",class_="css-d7j1kk")[0].text.split(" - ")[0]
        location_name=jobs[0].find_all("span",class_="css-5wys0k")[0].text.split(", ")[0]
        country=jobs[0].find_all("span",class_="css-5wys0k")[0].text.split(", ")[1]
        job_type=jobs[0].find_all("span",class_="css-1ve4b75 eoyjyou0")[0].text
        exe_level=jobs[0].find_all("div",class_="css-1lh32fc")[0].next_sibling.text.split(" · ")[0]
        exe_year=jobs[0].find_all("div",class_="css-1lh32fc")[0].next_sibling.text.split(" · ")[1]
        skills= "".join (jobs[0].find_all("div",class_="css-1lh32fc")[0].next_sibling.text.split(" · ")[2:])
        df.loc[len(df.index)]=[job_name, company_name, country, location_name, job_type, exe_level, exe_year, skills]
     
```
